```python
import numpy as np
import pandas as pd
salary_df=pd.read_csv('Salary_Data.csv')
salary_df
```

```python
salary_df.isnull().sum()
```

```python
salary_df['Salary']=salary_df['Salary'].fillna(salary_df['Salary'].mean())
```

```python
salary_df=salary_df.dropna(axis='index',how='any')
```

```python
salary_df.isnull().sum()
```

```python
salary_df.describe()
```

```python
salary_df.info()
```

```python
salary_df.to_csv('filtered_salary.csv',index=False)
```

```python
cars_df=pd.read_csv('used_cars_data.csv')
cars_df
```

```python
cars_df.isnull().sum()
```

```python
cars_df['Price']=cars_df['Price'].fillna(cars_df['Price'].median())
cars_df.drop(columns=['New_Price'],inplace=True)
cars_df.dropna(axis='index',how='any',subset=['Mileage','Engine','Power','Seats'],inplace=True)
```

```python
cars_df.isnull().sum()
```

```python
cars_df.info()
```

```python
cars_df.to_csv('filtered_cars.csv',index=False)
```

```python
import matplotlib.pyplot as plt
plt.hist(salary_df['Salary'])
plt.show()
```

```python
plt.boxplot(salary_df['Salary'])
plt.show()
```

```python
Q1=salary_df['Salary'].quantile(0.25)
Q3=salary_df['Salary'].quantile(0.75)
IQR=Q3-Q1
lowerLimit=Q1-1.5*IQR
upperLimit=Q3+1.5*IQR
outliers=salary_df[(salary_df['Salary']<lowerLimit)|(salary_df['Salary']>upperLimit)]
outliers
```

```python
cleaned_salary=salary_df[(salary_df['Salary']>=lowerLimit)&(salary_df['Salary']<=upperLimit)]
plt.boxplot(salary_df['Salary'])
plt.show()
```
