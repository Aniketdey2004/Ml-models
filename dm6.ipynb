```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
```

```python
data = {
    'Age': ['<=30','<=30','31-40','>40','>40','>40','31-40','<=30','<=30','>40','<=30','31-40','31-40','>40'],
    'Income': ['High','High','High','Medium','Low','Low','Low','Medium','Low','Medium','Medium','Medium','High','Medium'],
    'Student': ['No','No','No','No','Yes','Yes','Yes','No','Yes','Yes','Yes','No','Yes','No'],
    'Credit_rating': ['Fair','Excellent','Fair','Fair','Fair','Excellent','Excellent','Fair','Fair','Fair','Excellent','Excellent','Fair','Excellent'],
    'Buys_computer': ['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']
}

df = pd.DataFrame(data)
df
```

```python
# ============================================================
le = LabelEncoder()
for col in df.columns:
    df[col] = le.fit_transform(df[col])
df
```

```python
X = df.drop('Buys_computer', axis=1)
y = df['Buys_computer']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

```python
# ============================================================
model = DecisionTreeClassifier(criterion='entropy', random_state=42)
model.fit(X_train, y_train)

# ============================================================
# Step 6: Visualize the Decision Tree
# ============================================================
plt.figure(figsize=(12,8))
plot_tree(model, 
          feature_names=X.columns,
          class_names=['No', 'Yes'],
          filled=True,
          rounded=True)
plt.title("Decision Tree - Buys Computer")
plt.show()
```

```python
y_pred = model.predict(X_test)
print("Predictions:", y_pred, "\n")
```

```python
acc = accuracy_score(y_test, y_pred)
print(f"Accuracy: {acc:.2f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['No','Yes']))
```

```python
import pandas as pd
import numpy as np

# Step 1: Create Dataset-2
data = {
    'CustomerID': list(range(1, 51)),
    'Gender': ['Male','Female','Male','Female','Male','Female','Male','Female','Male','Female',
               'Male','Female','Male','Female','Male','Female','Male','Female','Male','Female',
               'Male','Female','Male','Female','Male','Female','Male','Female','Male','Female',
               'Male','Female','Male','Female','Male','Female','Male','Female','Male','Female',
               'Male','Female','Male','Female','Male','Female','Male','Female','Male','Female'],
    'CarType': ['Sedan','Hatchback','SUV','Sedan','Hatchback','SUV','Sedan','Hatchback','SUV','Sedan',
                'Hatchback','SUV','Sedan','Hatchback','SUV','Sedan','Hatchback','SUV','Sedan','Hatchback',
                'SUV','Sedan','Hatchback','SUV','Sedan','Hatchback','SUV','Sedan','Hatchback','SUV',
                'Sedan','Hatchback','SUV','Sedan','Hatchback','SUV','Sedan','Hatchback','SUV','Sedan',
                'Hatchback','SUV','Sedan','Hatchback','SUV','Sedan','Hatchback','SUV','Sedan','Hatchback'],
    'ShirtSize': ['M','L','XL','S','M','L','XL','S','M','L',
                  'XL','S','M','L','XL','S','M','L','XL','S',
                  'M','L','XL','S','M','L','XL','S','M','L',
                  'XL','S','M','L','XL','S','M','L','XL','S',
                  'M','L','XL','S','M','L','XL','S','M','L'],
    'Class': ['C0','C1','C0','C1','C1','C0','C1','C0','C1','C0',
              'C0','C1','C1','C0','C0','C1','C1','C0','C1','C0',
              'C0','C1','C1','C0','C1','C0','C1','C0','C0','C1',
              'C1','C0','C1','C0','C0','C1','C0','C1','C1','C0',
              'C0','C1','C1','C0','C1','C0','C1','C0','C1','C0']
}

df = pd.DataFrame(data)
print("=== Dataset-2 (first 10 rows) ===")
print(df.head(10), "\n")
```

```python
# Step 2: Define impurity functions
def gini_index(classes):
    p = np.array(classes) / np.sum(classes)
    return 1 - np.sum(p**2)

def entropy(classes):
    p = np.array(classes) / np.sum(classes)
    return -np.sum([pi * np.log2(pi) for pi in p if pi > 0])

def classification_error(classes):
    p = np.array(classes) / np.sum(classes)
    return 1 - np.max(p)
```

```python
class_counts = df['Class'].value_counts().to_dict()
print("Class distribution:", class_counts)
```

```python
values = list(class_counts.values())
print("\nOverall Gini Index:", round(gini_index(values), 4))
print("Overall Entropy:", round(entropy(values), 4))
print("Overall Classification Error:", round(classification_error(values), 4))
```

```python
# Step 4: Calculate metrics for each split (example: by Gender)
print("\n--- Split by Gender ---")
for gender, group in df.groupby('Gender'):
    counts = group['Class'].value_counts().to_dict()
    values = list(counts.values())
    print(f"{gender}: {counts}")
    print(f"  Gini = {round(gini_index(values),4)}, Entropy = {round(entropy(values),4)}, Error = {round(classification_error(values),4)}")
```

```python
# Step 5: Similarly by CarType
print("\n--- Split by CarType ---")
for car, group in df.groupby('CarType'):
    counts = group['Class'].value_counts().to_dict()
    values = list(counts.values())
    print(f"{car}: {counts}")
    print(f"  Gini = {round(gini_index(values),4)}, Entropy = {round(entropy(values),4)}, Error = {round(classification_error(values),4)}")
```
