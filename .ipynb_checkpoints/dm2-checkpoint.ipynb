```python
import pandas as pd
df=pd.read_csv('animal_characteristics.csv')
df
```

```python
print(df.shape)
```

```python
for col in df.columns:
    print(df[col].value_counts())
```

```python
def classifier(df):
    if df['Give Birth']=='no' and df['Can Fly']=='yes':
        return "Birds"
    elif df['Give Birth']=="no" and df['Live in Water']=='yes':
        return "Fishes"
    elif df['Give Birth']=='yes' and df['Blood Type']=='warm':
        return 'Mammals'
    elif df['Give Birth']=='no' and df['Can Fly']=='no':
        return "Reptiles"
    elif df['Live in Water']=='sometimes':
        return 'Amphibians'
    else:
        return 'Unknown'
```

```python
df['Class']=df.apply(classifier,axis=1)
df
```
